### Question 8-1. Please create a function that prints the contents of a group.

###  When you execute your function you should see the following results.

```python
def print_group(df,column):
    df_grouped = df.groupby(column)
    
    for each in df_grouped.groups.keys():
        print(each) 
        print(df.iloc[df_grouped.groups[each]].head())
        
print_group(adj_close_extracted,'Symbol')
```

> .size will tell us the count of items in each group

```python
adj_close_extracted_grouped.size()
```

#### Question 8-2. As you know, a specific group can be retrieved using .get_group() which returns a DataFrame representing the specified group. Please execute so that you can have the output below.

#### Question 11-4. Plot the adjusted close for EVERY day in 2008.

```python
plt.figure(figsize=(16,8))
plt.plot(dow.T[5], color = 'green')
plt.title('Adjusted Close Price for The Dow in 2008')
plt.xlabel('Date')
plt.ylabel('Adjusted Close')
plt.show()
plt.show()
```

#### Question 11-5. Now over-plot this plot with a `red dot` marker for every day where the dow volume was greater than 5.5 billion.

```python
red_plot=[]
for i in range(len(dow.T[5])):
    if i in array_index[0]:
        red_plot.append(dow.T[5][i])
    else:
        red_plot.append(np.nan)
red_plot

x_label = list(range(len(dow.T[5])))

plt.figure(figsize=(16,8))
plt.plot(x_label,dow.T[5], color = 'green')
plt.scatter(x_label,red_plot,color='r')
plt.title('Adjusted Close Price for The Dow in 2008')
plt.xlabel('Date')
plt.ylabel('Adjusted Close')
plt.show()
```
