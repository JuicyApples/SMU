### How to Use Python for Your Quant. Finance Learning Journey

#### You can simulate Geometric Brownian Motion :)
#### You can create your own financial world -- yay :)

$$
S_t = S_{t-1} e^{(\mu - \frac{\sigma^2}{2})\Delta t + \sigma \sqrt{\Delta t} Z}
$$

```python
drift = 0.04
vol = 0.50

# Parameters -

def create_financial_time_series(Rows,
                                 Columns,
                                 freq = "1 min"):

    
    Rows = int(Rows)
    Columns = int(Columns)
    
    INDEX =\
    (
    pd
    .date_range("2023-09-09",
                periods = Rows,
                freq = freq)
    )

    Timedelta = (INDEX[1] - INDEX[0]) / pd.Timedelta(value = "365D")
    
    COLUMNS = ["Our instrument %d"   % i for i in range(Columns)
              ]
    
    GB =\
    (
    np
    .exp(np.cumsum(
                   (drift - 0.5 * vol ** 2) * Timedelta + 
                    vol * np.sqrt(Timedelta) * np.random.standard_normal((Rows, Columns)),
                    axis = 0
                   )
         )
    )

    GB = GB / GB[0] * 300 # Normalization

# Now we are ready to product (generate) synthetic DF (tabular data)

    DF =\
    (
        pd
        .DataFrame(GB,
                   index = INDEX,
                   columns = COLUMNS)
    )
    
    return DF

# For Script Execution

if __name__ == "__main__":
    rows = 500
    columns = 18
    freq = "B"
```

```python
# Team, let's create a synthetic (simulated) financial time-series data

(
    create_financial_time_series(500,
                                 15,
                                 "B")
)
```

```python
(
    create_financial_time_series(500,
                                 15,
                                 "B")
).plot(figsize = [18, 10]
      )
```

### Regression Discontinuity `Assumption Ground`

$$
    \lim_{r \to c^-} E[Y_{ti} | R_i = r] = \lim_{r \to c^+} E[Y_{ti} | R_i = r]
$$

> If the above assumption holds true, we can find the causal effect at _`the threshold`_.

```python
alcohol.head(3)
```

```python
df_CLEANED =\
(
    alcohol
    .assign(threshold = (alcohol["agecell"] > 0).astype(int)
           )
)
```

```python
df_CLEANED
```

### MODEL for Regression Discontinuity

> formal expression of our causal model follows...

$$
y_i = \beta_0 + \beta_1 r_i + \beta_2 \mathcal{I}\{r_i > c\} + \beta_3 \mathcal{I}\{r_i > c\} \times r_i + \epsilon_i
$$

```python
# in statsmodels formula 
# weighted least squares

import statsmodels.formula.api as smf

Model =\
(
    smf
    .wls("all ~ agecell * threshold",
         data = df_CLEANED)
    .fit()
)
```

```python
Model \
    .summary() \
    .tables[1]
```

```python
Model.fittedvalues
```

```python
ax =\
(
    alcohol
    .plot
    .scatter(x = "agecell",
             y = "all")
)

(
    alcohol
    .assign(predictions = Model.fittedvalues)
    .plot(x = "agecell",
          y = "predictions",
          ax = ax,
          color = "red")
)

plt.title("Regression Discontinuity")
```

```python
(plt
    .figure(figsize = [10, 10]
           )
)

for p, cause in enumerate(["all", "suicide", "mva"], 1):
    
    ax = plt.subplot(3, 1, p)
    
    (    alcohol
        .plot
        .scatter(x = "agecell",
                 y = cause,
                 ax = ax)
    )

    models =\
    (        smf
        .wls(f"{cause} ~ agecell * threshold",
             data = df_CLEANED)
        .fit()
    )
    # ate = average treatment effects (the potential causal impact)
    ate_percent =\
    (100 * ((models.params["threshold"] + models.params["Intercept"]) / models.params["Intercept"] - 1)
    )
    
    (
    alcohol
        .assign(predictions = models.fittedvalues)
        .plot(x = "agecell",
              y = "predictions",
              ax = ax,
              color = "red"
             )
    )

    plt.title(f"Effects of Alcohol on Mortality: {np.round(ate_percent, 2)}%")
    
plt.tight_layout()
```

### Triangle `Kernel`

The triangle kernel \( K \) can be defined as:

$$
K(R, c, h) = \mathcal{I}\{|R-c| \leq h\} \times \left(1 - \frac{|R-c|}{h}\right)
$$


```python
def kernel(R, c, h):
    indicator = (np.abs(R - c) <= h).astype(float)
    return indicator * (1 - np.abs(R - c) / h)
```

```python
plt.plot(alcohol["agecell"],
         kernel(alcohol["agecell"],
                c = 0, h = 1)
        )

plt.ylabel("Kernel Weights")
plt.xlabel("Age (Centered)")

plt.title("Triangle Kernel Weights by Age (Centered)")
```

```python
Model_Correcting_Biases =\
(
    smf
    .wls("all ~ agecell * threshold",
         data = df_CLEANED,
         weights = kernel(alcohol["agecell"],
                          c = 0, 
                          h = 1)
        )
    .fit()
)

(Model
    .summary()
    .tables[1]
)
```

```python
(plt
    .figure(figsize = [10, 10]
           )
)

weights =\
(
    kernel(alcohol["agecell"],
           c = 0,
           h = 1
          )
)

for p, cause in enumerate(["all", "suicide", "mva"], 1):
    
    ax = plt.subplot(3, 1, p)
    
    alcohol.plot.scatter(x = "agecell",
                         y = cause,
                         ax = ax)

    models = smf.wls(f"{cause} ~ agecell * threshold",
                     data = df_CLEANED,
                     weights = weights).fit()

    # ate = average treatment effects (the potential causal impact)
    ate_percent =\
    (100 * ((models.params["threshold"] + models.params["Intercept"]) / models.params["Intercept"] - 1)
     )
    
    (alcohol
        .assign(predictions = models.fittedvalues)
        .plot(x = "agecell",
              y = "predictions",
              ax = ax,
              color = "red"
             )
    )

    plt.title(f"Effects of Alcohol on Mortality (After adjusting for biases): {np.round(ate_percent, 2)}%")
    
plt.tight_layout()
```
