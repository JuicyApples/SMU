# Returns

```r
# Define the window size
daily_returns <- df$Daily.Return
window_size <- 7025

first_period_fit <- NULL

# Initialize vectors to store predictions
volatility_predictions <- numeric(length(daily_returns) - window_size)

# Loop through the data using the defined window size
for(i in 1:(length(daily_returns) - window_size)) {
  # Subset the data for the current window
  window_data <- daily_returns[i:(i + window_size - 1)]
  
  # Define the GARCH model (e.g., GARCH(1,1))
  spec <- ugarchspec(variance.model = list(model ="sGARCH",garchOrder = c(1, 1)), 
                     mean.model = list(armaOrder = c(0, 0), include.mean = TRUE), 
                     distribution.model = "norm")
  
  # Fit the GARCH model to the window data
  fit <- ugarchfit(spec = spec, data = window_data)

  if(i == 1) {
    first_period_fit <- fit
      }

  forecast <- ugarchforecast(fit, n.ahead = 1)
  # Forecast the next period's volatility
  volatility_predictions[i] <- forecast@forecast$sigmaFor
}
print(volatility_predictions)
```

```r
plot(volatility_predictions, type="l", col="blue", lty=1, pch=16, ylab="Volatility Predictions", main="Volatility Predictions (RV)")
```
